```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("dataset.csv")
print(dataset)
```

```python
X = dataset.iloc[:,3:5]
print(X)
```

```python
from sklearn.cluster import KMeans
kmeans= KMeans( n_clusters = 3 , init = 'k-means++')
kmeans.fit(X)

print(kmeans.cluster_centers_)
```

```python
results = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init='k-means++',random_state=123)
    kmeans.fit(X)
    results.append(kmeans.inertia_)

plt.plot(range(1,11),results)
```

```python
kmeans = KMeans (n_clusters = 4, init='k-means++', random_state= 123)
Y_pred = kmeans.fit_predict(X)
import numpy as np
print(X)
X_np = X.to_numpy()  # convert DataFrame to NumPy array

plt.scatter(X_np[Y_pred == 0, 0], X_np[Y_pred == 0, 1], s=100, c='red')
plt.scatter(X_np[Y_pred == 1, 0], X_np[Y_pred == 1, 1], s=100, c='blue')
plt.scatter(X_np[Y_pred == 2, 0], X_np[Y_pred == 2, 1], s=100, c='purple')
plt.scatter(X_np[Y_pred == 3, 0], X_np[Y_pred == 3, 1], s=100, c='black')
plt.title("KMeans Clustering Results")
plt.show()
```
