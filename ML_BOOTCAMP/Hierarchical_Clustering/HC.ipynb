```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('dataset.csv')
print(dataset)
```

```python
X = dataset.iloc[:,3:]
print(X)
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters = 4, linkage = 'ward')
Y_pred = ac.fit_predict(X)
print(Y_pred)
```

```python
import numpy as np
print(X)
X_np = X.to_numpy()  # convert DataFrame to NumPy array

plt.scatter(X_np[Y_pred == 0, 0], X_np[Y_pred == 0, 1], s=100, c='red')
plt.scatter(X_np[Y_pred == 1, 0], X_np[Y_pred == 1, 1], s=100, c='blue')
plt.scatter(X_np[Y_pred == 2, 0], X_np[Y_pred == 2, 1], s=100, c='purple')
plt.scatter(X_np[Y_pred == 3, 0], X_np[Y_pred == 3, 1], s=100, c='black')
plt.title("Agglomerative Clustering Results")
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()
```
