```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('odev_tenis.csv')
print(dataset)

c = pd.get_dummies(dataset.iloc[:, :1])
print(c)

from sklearn.preprocessing import OneHotEncoder, LabelEncoder


# OneHotEncoder: outlook
ohe = OneHotEncoder(sparse_output=False)
outlook_encoded = ohe.fit_transform(dataset[['outlook']])
outlook_df = pd.DataFrame(outlook_encoded, columns=ohe.get_feature_names_out(['outlook']))

# Dataset
dataset = pd.concat([outlook_df, dataset.drop('outlook', axis=1)], axis=1)
from sklearn.preprocessing import LabelEncoder

le_play = LabelEncoder()
dataset['play'] = le_play.fit_transform(dataset['play'])

le_windy = LabelEncoder()
w = le_windy.fit_transform(dataset['windy'])
dataset['windy'] = w 
print(dataset)

y = pd.DataFrame(w, columns=['windy'])
print(y)

 

print(y)
X = dataset.drop(['windy'], axis=1)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=0)

print(x_train)
print(y_train)
print(x_test)
print(y_test)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)


y_pred = regressor.predict(x_test)
print("\n--- Sklearn Linear Regression Tahminleri ---")
print(y_pred)
print(x_test)
```

```python
import statsmodels.api as sm

# Backward elimination başlat
# X'e sabit ekleyelim
X_be = sm.add_constant(X)

# İlk OLS fit
model = sm.OLS(y, X_be).fit()
print(model.summary())

# Backward elimination döngüsü
def backward_elimination(X, y, sl=0.05):
    X_model = X.copy()
    while True:
        X_with_const = sm.add_constant(X_model)
        model = sm.OLS(y, X_with_const).fit()
        pvalues = model.pvalues.iloc[1:]  # sabiti at
        max_pval = pvalues.max()
        if max_pval > sl:
            excluded_var = pvalues.idxmax()
            print(f"Elendi: {excluded_var} (p-value={max_pval:.4f})")
            X_model = X_model.drop(columns=excluded_var)
        else:
            break
    return X_model, model

# Backward elimination uygula
X_selected, final_model = backward_elimination(X, y)

# Son model summary
print("\n--- Final OLS Summary ---")
print(final_model.summary())

# Son seçilen değişkenler
print("\n--- Seçilen değişkenler ---")
print(X_selected.columns.tolist())
```
