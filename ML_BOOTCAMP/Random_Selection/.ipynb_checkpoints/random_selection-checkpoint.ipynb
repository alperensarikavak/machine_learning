```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
dataset = pd.read_csv('dataset.csv')
print(dataset)
```

```python
n = 9999 #rows
d = 10 #columns
sum = 0
selected = []
for i in range(0,n):
    ad = random.randrange(d)
    selected.append(ad)
    prize = dataset.values[i,ad]
    sum = sum + prize

```

```python
from collections import Counter

counts = Counter(selected) 

ads = list(range(10))
click_counts = [counts.get(i, 0) for i in ads]

plt.bar(ads, click_counts)
plt.title("Ad Selection Count (Random Selection)")
plt.xlabel("Ad Number")
plt.ylabel("Selection Count")
plt.xticks(ads)
plt.show()
print(click_counts)
```
